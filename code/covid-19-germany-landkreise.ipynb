```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('RKI_COVID19.csv', parse_dates=['Meldedatum','Datenstand','Refdatum'])
```

```python
df.head()
```

```python
Tue = df[df['Landkreis']=='LK Tübingen'] #Define Tue variable by filtering the relevant rows and columns in the dataframe
```

```python
Tue.head()
```

```python
#drop all columns that we don't care
Tue = Tue.drop(columns=['ObjectId','Bundesland','Altersgruppe','IdBundesland','IdLandkreis','IstErkrankungsbeginn','NeuGenesen', 'Altersgruppe2'])
```

```python
Tue.head()
```

```python
Tue = Tue.sort_values("Meldedatum") #sort the rows as increasing according to the Meldedatum
```

```python
Total_Cases = Tue['AnzahlFall'].cumsum() #calculate cummulative sum for each entry
Total_Deaths = Tue['AnzahlTodesfall'].cumsum() 
```

```python
SMALL_SIZE = 16
MEDIUM_SIZE = 20
BIGGER_SIZE = 28

plt.rc('font', size=SMALL_SIZE)          # controls default text sizes
plt.rc('axes', titlesize=SMALL_SIZE)     # fontsize of the axes title
plt.rc('axes', labelsize=MEDIUM_SIZE)    # fontsize of the x and y labels
plt.rc('xtick', labelsize=SMALL_SIZE)    # fontsize of the tick labels
plt.rc('ytick', labelsize=SMALL_SIZE)    # fontsize of the tick labels
plt.rc('legend', fontsize=SMALL_SIZE)    # legend fontsize
plt.rc('figure', titlesize=BIGGER_SIZE)  # fontsize of the figure title
```

```python
#prepare and plot the figure
fig, ax, = plt.subplots()
ax.plot(Tue['Meldedatum'],Total_Cases, color='b')
fig.suptitle('The Reported Spread of COVID19 in Tübingen')
ax.set_xlabel('Time')
ax.set_ylabel('Number of cases', color='b')
ax2 = ax.twinx()
ax2.plot(Tue['Meldedatum'], Total_Deaths, color='r')
ax2.set_ylabel('Number of deaths', color='r')
fig.set_size_inches([14,10])
plt.show()
```

```python
#save the figure
fig.savefig('COVID19_DE_Tuebingen2.png')
```
